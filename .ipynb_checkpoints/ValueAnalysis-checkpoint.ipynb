```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('datasets/allstats.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Date'] = df['Date'].astype(str)
df['Date'] = pd.to_datetime(df['Date'], yearfirst = True)
```

```python
df['Year'] = pd.DatetimeIndex(df['Date']).year
```

```python
df['Player'] = df['Player'].astype(str)
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
yearmask = df['Year'] == 2019
players = df[yearmask]
players.head()
```

```python
fantstats = players[['Player','FantasyPoints']].groupby(['Player']).sum()
```

```python
fantstats.head()
```

```python
fantstats.sort_values(by=['FantasyPoints'], ascending = False)
```

```python
fantstats = fantstats.merge(df[['Player','Pos']], on = 'Player')
```

```python
fantstats.drop_duplicates(inplace = True)
```

```python
fantstats.sort_values(by = ['FantasyPoints'], ascending = False)
```

```python
from requests import get
from bs4 import BeautifulSoup
```

```python
url = 'https://www.fantasypros.com/nfl/adp/half-point-ppr-overall.php'
response = get(url)
soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find('table', {'id': 'data'})
df = pd.read_html(str(table))[0]
df.head()
```

```python
df.dtypes
```

```python
df['Player Team (Bye)'] = df['Player Team (Bye)'].astype(str)
```

```python
df['Player'] = df['Player Team (Bye)'].apply(lambda x: x.split(' ')[:2]).apply(' '.join)
df.head()
```

```python
import re 
  
def remove(list): 
    pattern = '[0-9]'
    list = [re.sub(pattern, '', i) for i in list] 
    return list
  
# Driver code  
  
list = ['4geeks', '3for', '4geeks'] 
print(remove(list)) 
```

```python
df['Position'] = [re.sub('[0-9]', '',i) for i in df['POS']]
df.head()
```

```python
print(df['Position'].value_counts())
print(df.shape)
```

```python
master = fantstats.merge(df[['Player', 'Rank', 'Position', 'AVG']], on = 'Player' )
master.sort_values(by = ['AVG'], inplace = True)
master.drop_duplicates(subset ="Player", keep = False, inplace = True)
no_k = master['Position'] != 'K'
master = master[no_k]
master = master[:200]
master.head(10)
#master.tail()
```

```python
master['Position'].value_counts()
```

```python
pos = ['RB', 'WR', 'QB', 'TE']
for pos in pos:
    posmask = master['Position'].str.contains(pos)
    if pos == 'RB':
        RBdata = master[posmask]
        RBdata.name = '{pos}'.format(pos = pos)
    elif pos == 'WR':
        WRdata = master[posmask]
        WRdata.name = '{pos}'.format(pos = pos)
    elif pos == 'QB':
        QBdata = master[posmask]
        QBdata.name = '{pos}'.format(pos = pos)
    elif pos == 'TE':
        TEdata = master[posmask]
        TEdata.name = '{pos}'.format(pos = pos)
```

```python
RBdata
```

```python
Tables = [RBdata, WRdata, QBdata, TEdata]
for table in Tables:
    avg = table['FantasyPoints'].mean()
    table['Rank'] = np.arange(1,len(table)+1).astype(float)
    table['VOA'] = ((table['FantasyPoints'])/ avg) 
    table['PosValue'] = (table['VOA'])/((len(table)+1-table['Rank'])/(len(table)))
RBdata
```

```python
master = pd.concat(Tables, join = 'outer', ignore_index = False, sort = False)
master
```

```python
master.set_index(np.arange(0,len(master)), inplace = True)
master['Value'] = (master['VOA'])/((len(master)+1-master.index)/len(master))

master
```
