# Fantasy Pros projections webscrape. Projections from 2019

```python
pos = ['qb', 'rb', 'wr', 'te']

urls = {
    
}

for pos in pos:
    urls.update({'{pos}projections'.format(pos = pos) : 'https://www.fantasypros.com/nfl/projections/{pos}.php?week=draft'.format(pos = pos)})
    
dfs = []
```

```python
for key, url in urls.items():
    response = get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table', {'id': 'data'})
    df = pd.read_html(str(table))[0]
    dfs.append(df)
```

```python
columns = ['Fum', 'FPTS', 'PassingAtt', 'Completions', 'Int', 'PassingTds', 'PassingYds', 'Rec', 'ReceivingTds', 'ReceivingYds', 'RushingAtt', 'RushingTds', 'RushingYds', 'Player']
```

```python
fantasy_pros = pd.concat(dfs, join = 'outer', ignore_index = False, sort = False)
fantasy_pros.fillna(0, inplace = True)
fantasy_pros.columns = df.columns.droplevel(level=0)
fantasy_pros.columns = columns
fantasy_pros['Player'] = fantasy_pros['Player'].apply(lambda x: x.split(' ')[:2]).apply(' '.join)
fantasy_pros.sort_values(by = ['FPTS'], ascending = False)
fantasy_pros.head()
```

```python
fantasy_pros['FantasyPoints'] = fantasy_pros['PassingYds']/25 + fantasy_pros['PassingTds']*4 - fantasy_pros['Int']*2 - fantasy_pros['Fum']*2 + 0.5* fantasy_pros['Rec'] + fantasy_pros['ReceivingYds']/10 + fantasy_pros['ReceivingTds']*6 + fantasy_pros['RushingYds']/10 + fantasy_pros['RushingTds']*6
```

```python
dfs = []
for a in range(2,6):
    for b in range(1,3):
        temp = []
        rows = driver.find_elements_by_xpath("""/html/body/center/table[2]/tbody/tr[2]/td[1]/table[6]/tbody/tr/td/table/tbody/tr""")
        for tr in range(2,len(rows)+1):
            cols = driver.find_elements_by_xpath("""/html/body/center/table[2]/tbody/tr[2]/td[1]/table[6]/tbody/tr/td/table/tbody/tr[2]/td""")
            temp2 = []
            for td in range(2,len(cols)+1):
                elem = driver.find_element_by_xpath("""/html/body/center/table[2]/tbody/tr[2]/td[1]/table[6]/tbody/tr/td/table/tbody/tr[{tr}]/td[{td}]""".format(tr = tr, td = td))
                temp2.append(elem.text)
            temp.append(temp2)
        df = pd.DataFrame(temp[1:], columns = temp[0])
        dfs.append(df)
        driver.find_element_by_xpath("""/html/body/center/table[2]/tbody/tr[2]/td[1]/div/a[1]""").click()
    driver.find_element_by_xpath("""/html/body/center/table[2]/tbody/tr[2]/td[1]/table[4]/tbody/tr/td/a[{a}]""".format(a = a)).click()

print(dfs, len(dfs))
```

# NFL Fantasy Football Projections (2020)

```python
url = 'https://fantasy.nfl.com/research/projections?position=O&sort=projectedPts&statCategory=projectedStats&statSeason=2019&statType=seasonProjectedStats'
url2 = 'https://fantasy.nfl.com/research/projections?offset=26&position=O&sort=projectedPts&statCategory=projectedStats&statSeason=2019&statType=seasonProjectedStats&statWeek=17'
url3 = 'https://fantasy.nfl.com/research/projections?offset=51&position=O&sort=projectedPts&statCategory=projectedStats&statSeason=2019&statType=seasonProjectedStats&statWeek=17'
url4 = 'https://fantasy.nfl.com/research/projections?offset=76&position=O&sort=projectedPts&statCategory=projectedStats&statSeason=2019&statType=seasonProjectedStats&statWeek=17'
url5 = 'https://fantasy.nfl.com/research/projections?offset=101&position=O&sort=projectedPts&statCategory=projectedStats&statSeason=2019&statType=seasonProjectedStats&statWeek=17'
url6 = 'https://fantasy.nfl.com/research/projections?offset=126&position=O&sort=projectedPts&statCategory=projectedStats&statSeason=2019&statType=seasonProjectedStats&statWeek=17'
url7 = 'https://fantasy.nfl.com/research/projections?offset=151&position=O&sort=projectedPts&statCategory=projectedStats&statSeason=2019&statType=seasonProjectedStats&statWeek=17'
url8 = 'https://fantasy.nfl.com/research/projections?offset=176&position=O&sort=projectedPts&statCategory=projectedStats&statSeason=2019&statType=seasonProjectedStats&statWeek=17'
url9 = 'https://fantasy.nfl.com/research/projections?offset=201&position=O&sort=projectedPts&statCategory=projectedStats&statSeason=2019&statType=seasonProjectedStats&statWeek=17'

urls = [url, url2, url3, url4, url5, url6, url7, url8, url9]
data = []
```

```python
for links in urls:
    response = get(links)
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find_all('table', {'class': 'tableType-player hasGroups'})
    df = pd.read_html(str(table))[0]
    df.columns = df.columns.droplevel(level=0)
    data.append(df)

len(data)
```

```python
df = pd.concat(data, join = 'outer', sort = False)
df.fillna(0, inplace = True)
df.reset_index(inplace = True)
df
```
