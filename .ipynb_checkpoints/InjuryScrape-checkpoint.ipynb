```python
dfs = []
for key, url in urlDict.items():  
    chrome_path = r"C:\Users\cjohn\OneDrive\Documents\chromedriver.exe"
    driver = webdriver.Chrome(chrome_path)
    driver.get(url)
    for div in range(4,68):
        if div % 2 != 0:
            temp = []
            rows = driver.find_elements_by_xpath("""//*[@id="leftcol"]/div[{div}]/div""".format(div = div))
            for tr in range(2,len(rows)+1):
                temp2 = []
                b = div - 1
                team = driver.find_element_by_xpath("""//*[@id="leftcol"]/div[{b}]/b""".format(b = b))
                temp2.append(team.text)
                week = driver.find_element_by_xpath("""//*[@id="dropdownMenuInj"]""")
                temp2.append(week.text)
                year = driver.find_element_by_xpath("""//*[@id="dropdownMenuYear"]""")
                temp2.append(year.text)
                position = driver.find_element_by_xpath("""//*[@id="leftcol"]/div[{div}]/div[{tr}]/div[1]""".format(div = div, tr = tr))
                temp2.append([position.text])
                cols = driver.find_elements_by_xpath("""//*[@id="leftcol"]/div[5]/div[1]/div""")
                for td in range(2,len(cols)+1):
                    elem = driver.find_element_by_xpath("""//*[@id="leftcol"]/div[{div}]/div[{tr}]/div[{td}]""".format(td = td, tr = tr, div = div))
                    temp2.append(elem.text)
                temp.append(temp2)
            df = pd.DataFrame(temp, columns = ['Team', 'Week', 'Year', 'Player', 'Injury', 'Wed', 'Thu', 'Fri', 'Game Status'])
            dfs.append(df)
    driver.quit()
```

```python
dfs[0]
```

```python
df = pd.concat(dfs, join = 'outer', ignore_index = True)
df
```

```python
df['Player'] = df['Player'].apply(pd.Series)
df[['Player', 'Position']] = df['Player'].str.split(',', expand = True, n = 2)
df.head()
```

```python
ex = df['Game Status'].apply(lambda x: x.split(' vs')[0]).apply(lambda x: x.split('@')[0]).apply(lambda x: x.split(' ')[-1])
ex
```
