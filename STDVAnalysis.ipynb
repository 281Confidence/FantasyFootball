```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
adp = pd.read_csv('datasets/ADP.csv')
stats = pd.read_csv('datasets/allstats.csv')
```

```python
stats.dtypes
```

```python
stats['Date'] = stats['Date'].astype(str)
stats['Date'] = pd.to_datetime(stats["Date"], yearfirst = True)
stats
```

```python
stats['Year'] = pd.DatetimeIndex(stats['Date']).year
stats.dtypes
```

```python
yearmask = stats['Year'] == 2019
df = stats[yearmask].copy()
df.head()
```

```python
table = pd.pivot_table(df, values = 'FantasyPoints', index = ['Player'], columns = ['Week'])
table
```

```python
table['STDev'] = table.std(axis = 1)
table
```

```python
table.reset_index(inplace = True)
table
```

```python
table.drop(['index', 'level_0'], axis = 1, inplace = True)
table
```

```python
adp.drop(['Unnamed: 0'], axis = 1, inplace = True)
adp['Player Team (Bye)'] =adp['Player Team (Bye)'].astype(str)
adp['Player'] =adp['Player Team (Bye)'].apply(lambda x: x.split(' ')[:2]).apply(' '.join)
adp.head()
```

```python
adp.drop(['Player Team (Bye)', 'Rank'], axis = 1, inplace = True)
adp.head()
```

```python
master = adp.merge(table[['Player', 'STDev']], on = 'Player', )
master
```
